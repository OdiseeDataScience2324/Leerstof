### Data cleaning

Het doel van de bovenstaande vragen was om vertrouwd te geraken met de data die beschikbaar is in de dataset en met een aantal operaties en functies die vaak uitgevoerd worden op dataframes/datasets. 
In de volgende oefening gaan we een aantal stappen van het data-cleaning proces overlopen en uitvoeren. 

Transformeer het dataframe (hernoem het naar df_transformed) om aan de volgende vereisten te voldoen:
* Zorg ervoor dat er enkel rekening gehouden wordt met gereleased films.
* Houd de kolom met de gemiddelde_stem van elke film over
* Bereken een target kolom dat het winst percentage bevat: (revenue - budget) / budget. Let er hierop dat dat films met een niet bestaand buget (0, NA, NULL) niet meegeteld worden
* Maak een ordinal encoding van de originele taal
* Maak een multi-hot encoding van de verschillende genres. Let hierbij op dat sommige genres uit meerdere woorden bestaan.
* Maak een encoded kolom aan van de runtime waarbij:
  * de films met een runtime van 0 minuten -> de waarde unknown
  * de films met een runtime <= 30 minuten -> de waarde kortspeelfilm
  * de films met een runtime > 30 minuten -> de waarde langspeelfilm
* Bereken de volgende aantallen en plaats deze elk in een aparte kolom:
  * Het aantal woorden in de tagline
  * Het aantal leden in de cast (Maak hier een beste inschatting aangezien het moeilijk kan zijn dit exact te weten)
  * Het aantal crew-leden
* Alle kolommen die niet hierboven vermeld zijn moeten weggelaten worden.

Beschrijf hieronder de nodige stappen om elk punt uit te werken. Vermeld ook als je extra zaken moet toevoegen, zoals null waarden invullen:

**Antwoord:**
...

```python
# stap 1/2/3
df_origin_with_budget = df_movies_released[df_movies_released.budget > 0]
df = df_origin_with_budget[["budget", "revenue", "vote_average"]].copy()

# stap 3
df["winst"] = (df.revenue - df.budget) / df.budget

# stap 4
df["language_ord_enc"] = pd.factorize(df_origin_with_budget.original_language)[0]

# stap 5
one_hots = df_origin_with_budget['genres'].str.get_dummies(sep=' ')
df = pd.concat([df, one_hots], axis=1)
df["Science Fiction"] = df.Science
df.drop(["Science", "Fiction"], axis=1, inplace=True)

# stap 6
def enc_runtime(value):
    if value == 0:
        return "unknown"
    elif value <= 30:
        return "kortspeelfilm"
    else:
        return "langspeelfilm"
df["runtime_enc"] = df_origin_with_budget.runtime.apply(enc_runtime)

# stap 7a,b,c
df_origin_with_budget.tagline = df_origin_with_budget.tagline.fillna('')
df_origin_with_budget.cast = df_origin_with_budget.cast.fillna('')
df_origin_with_budget.crew = df_origin_with_budget.crew.fillna('')
df["tagline_num_words"] = df_origin_with_budget.tagline.str.split(' ').apply(len)
df["cast_size"] = df_origin_with_budget.cast.str.split(' ').apply(len)
df["crew_size"] = df_origin_with_budget.crew.str.split('},').apply(len)

#stap 8
df = df.drop(["budget", "revenue"], axis=1)

display(df)
```

Dit dataframe kan dan gebruikt worden voor analyse of machine learning modellen op te trainen.
Voor dat we verder gaan met de tussenstap, bewaar dit dataframe eerst als csv onder de naam **movies_transformed.csv**.

```python
df.to_csv("movies-dataset/movies_transformed.csv")
```
